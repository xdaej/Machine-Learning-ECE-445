**Problem 2:**

The plot is very representative of an exponential growth function. This makes sense as the plot is of GDP growth over time. The exponential growth function is in the form of: 

f(x) = A(1+r)^t

where A = initial amount $, r = rate in decimal, t = time passed

**Problem 3:**

```python
x2 = np.arange(0,90,10)
y2 = 1.05*(1+0.037)**x2
plt.figure(figsize=(10,8))
plt.plot(x2,y2, 'b-', label = 'Hypothesized GDP', marker = 'o')
plt.plot(x2,y, 'r-', label = 'Real GDP', marker = 'o')
plt.title('Hypothesized GDP function vs. Real GDP Growth', fontsize = 16)
plt.xlabel('Years passed', fontsize = 12)
plt.ylabel('Trillions of $', fontsize = 12)
plt.legend()

plt.show()
```

When plotting this graph it is crucial to recreate the x-values from date year to years passed. This ensures that the growth function is correctly retrieved from start of 0 years passed. 

Initial amount A was given as 1.05 trillion, while t was equal to years passed x2. R was experimented with, however it was found through quick googling that the general economic growth rate is ussually ~4%.

Thus R was experimented around 3-4% to obtain a pretty precise growth rate at 0.037 or 3.7%. **The graphs are overlaid to show the accuracy of the 3.7% growth rate.**
