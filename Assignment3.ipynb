** Gradient Descent**

Gradient Descent is an algorithm to find a minimum of a function.

Bivariate function takes in two variables.
For two variables, calculate the learning rate by partially deriving the first variable and then the second variable

Start at random point and move in negative direction of the gradient function to reach the local/global min.

```python
import numpy as np
#For bolding print statements to make easier for reader
B= '\033[1m'
E= '\033[0m'
```

Bivariate Function: f(w) = (w1^2 + w2-11)^2+(w1+w2^2-7)^2

substituting w1,w2 with x,y

Partial Derivatives =

dx: 2(2x(x^2+y-11)+x+y^2-7)

dy: 2(x^2+2y(x+y^2-7)+y-11)

Problem 1

```python
#algorithm starts at w0 = [0 -4].T
w0 = np.array([0,-4]).T

#precision, treshold to stop algorithm
precision = 10**-12

#max iterations = 10,000 before declaring failure
max_iters = 10000

#gradients as partial derivatives dx dy
dfx = lambda x,y: 2*(2*x*(x**2+y-11)+x+y**2-7)
dfy = lambda x,y: 2*(x**2+2*y*(x+y**2-7)+y-11)
```

```python
#step size rate of 0.005
rate = 0.005

#initialize iteration counter and step limits
previous_step_sizex = 1
previous_step_sizey = 1
iters = 0

#initialize w0[x,y] into x, y vars for looping w/o changing initial w0
x = w0[0]
y = w0[1]

#end loop when diff between 2 consecutive iterations are < 10^-12 precision
#or end if # iters > 10,000 max_iterations
while previous_step_sizex and previous_step_sizey> precision and iters < max_iters:
    prev_wx = x #store x's into prev_wx
    prev_wy = y #store y's into prev_wy
    
    #gradient descents of dx dy to update new x y
    x = x - rate*dfx(prev_wx,prev_wy)
    y = y - rate*dfy(prev_wx,prev_wy)

    #change diff of x and y for w[x,y] to update step limits for x y
    previous_step_sizex = abs(x - prev_wx)
    previous_step_sizey = abs(y - prev_wy)
    
    #iterate counter to update iteration limits
    iters = iters+1
    
    #print iterations
    print(B+"interation"+E,iters,B+"\nx,y values:"+E, "[",x,y,"]")
      
print(B+"\nWith", iters,"iterations"+E)
print(B+"The local min converges at:"+E, "[",x,y,"]")
```

**The convergence speed becomes faster with higher step size rate.** This is because the larger the step size, the larger the space between convergence is iterated through. More space covered per iteration means the faster the convergence is discovered.

The drawback to bigger step size however is having lower accuracy. Since more space is being jumped over, the values in the space aren't being as exact as a smaller step size jump.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg as spla
import scipy.stats
from scipy.stats import multivariate_normal as mn
#For bolding print statements to make easier for reader
B= '\033[1m'
E= '\033[0m'
```

```python
mean = np.array([-1, 0, 4, 1, 0.5]).T
I = np.identity(5)
cov = 2*I
```

```python
# Scipy's multivariate_normal function was hard to understand and the wrong results kept appearing.
# So I created my own multivariate_normal_fn using similar params

# n_samples -> N realizations

def multivariate_normal_fn(mean, covariance, n_samples=1):
    L = spla.cholesky(covariance)
    Z = np.random.normal(size=(n_samples,covariance.shape[0]))
    return Z.dot(L) + mean

N = [] # storing n realizations for log plot later
X = [] # storing the mean vectors into X

#Loop i => 1:6 for N^i realizations 
for i in range(0,6):
    n = 10**(i+1)
    N.append(n)
    x = multivariate_normal_fn(mean, cov, n_samples=n)
    X.append(x.mean(0))
    print(B+'Mean vector for 10^',i+1,'realizations:'+E)
    print(x.mean(0))

#Loop i => 1:6 errors
Error = [] #storing error vals for log plot later
for i in range(0,6):
    e = np.linalg.norm(X[i]-mean)
    Error.append(e**2)
    print(B+'Instantaneous Error for 10^',i+1,'realizations:'+E)
    print(e)
```

```python
#Error vs N as a log-log plot
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
line = ax.plot(Error,N)
ax.set_yscale('log')
ax.set_xscale('log')
fig.suptitle('Error vs. N as log-log')
plt.xlabel('Instantaneous Error')
plt.ylabel('N Realizations')
plt.show()
```

Given the log-log plot above, it is shown that the Instantaneous Error is inverse linearly related to N-realizations.

This means that the higher the realizations/samples provided, the smaller the Error becomes.

This makes sense as more data analyzed (in a large log scale) tends to decrease the error/variation amongst the data.
